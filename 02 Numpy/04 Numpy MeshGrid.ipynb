### Example 01

```python
import numpy as np
import matplotlib.pyplot as plt

xvalues = np.array([1,2,3,4])
yvalues = np.array([1,2,3,4])

xx,yy = np.meshgrid(xvalues,yvalues)

print("xx:",xx)
print("\n")
print("yy:",yy)

plt.plot(xx, yy, marker='.', color='k', linestyle='none')
```
