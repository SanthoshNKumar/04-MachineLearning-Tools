```python
#  From Dictionary

import pandas as pd
import numpy as np

dict = {"Name":["Santhosh","Anand","Sreedevi",'Prashanth','Sachin'],
        "Age":[32,34,31,30,35],
        "Sex":["Male","Male","Female","Male","Male"],
        "Company":["Microsft","Google","Facebook","Tesla","Volvo"],
        "Grade":["P1","P2","P1","P3","P1"]}

df = pd.DataFrame(dict)

df
```

```python
#7 Creating DataFrame using zip() function.

name = ['Santhosh','Anand','Sreedevi','Prashanth','Sachin']
age =[32,34,31,30,37]
company = ["Microsoft",'Google','Facebook','Tesla','Volvo']

list_of_tuples = list(zip(name, age,company))

columns = ['Name','Age','Company']

pd.DataFrame(list_of_tuples,columns = columns) 
```

```python
#13 Read tables from an HTML page using the pandas library built in read_html

import pandas as pd
import requests

url = 'https://www.goodcarbadcar.net/2020-us-vehicle-sales-figures-by-brand'

r = requests.get(url)

#if the response status is OK (200)
if r.status_code == 200:
    car_data_tables = pd.read_html(r.text)
    
# display the first table
car_data_tables[0]
```
