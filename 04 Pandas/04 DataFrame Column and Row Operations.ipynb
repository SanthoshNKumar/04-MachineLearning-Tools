```python
##  view duplicate rows

data = [{'Name': 'Santhosh', 'Age':32, 'Company':'Microsft'},
        {'Name': 'Anand', 'Age':34, 'Company':'Google'},
        {'Name': 'Sreedevi', 'Age':31, 'Company':'Facebook'},
        {'Name': 'Prashanth', 'Age':30, 'Company':'Testla'},
        {'Name': 'Sachin', 'Age':35, 'Company':'Volvo'},
        {'Name': 'Anand', 'Age':34, 'Company':'Google'}]

df = pd.DataFrame(data)

df
```

```python
duplicate = df[df.duplicated()]

duplicate
```

```python
# Removing duplicate rows

# dropping ALL duplicate values (keep = False)
# drop duplicates but keep one last entry (keep='last')
df.drop_duplicates(keep = 'last', inplace = True)

df
```

```python
## Add rows to Existing Data Frame
df.append({'Name': "Harsha","Age":32,"Company":"Philips"}, ignore_index=True)
```
