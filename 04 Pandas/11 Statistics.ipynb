```python
# absolute
   
df = pd.DataFrame({"A":[14, -4, 5, -4, 1], 
                   "B":[5, -2, 54, 3, -2],
                   "C":[-20, 20, 7, 3, 8],
                   "D":[-14, 3, 6, 2, -6]}) 
print(df.abs())
```

```python
import pandas as pd 
   
df = pd.DataFrame({"A":[14, 4, 5, 4, 1], 
                   "B":[5, 2, 54, 3, 2],
                   "C":[20, 20, 7, 3, 8],
                   "D":[14, 3, 6, 2, 6]}) 
print(df.prod())
```

```python
# Cumulative total of values.
import pandas as pd 
   
df = pd.DataFrame({"A":[14, 4, 5, 4, 1], 
                   "B":[5, 2, 54, 3, 2],
                   "C":[20, 20, 7, 3, 8],
                   "D":[14, 3, 6, 2, 6]}) 
print(df.cumsum(1))
```

```python
# Cumulative product of values.

import pandas as pd 
   
df = pd.DataFrame({"A":[14, 4, 5, 4, 1], 
                   "B":[5, 2, 54, 3, 2],
                   "C":[20, 20, 7, 3, 8],
                   "D":[14, 3, 6, 2, 6]}) 
print(df.cumprod())
```

```python
# Find Correlation
df['Age'].corr(df['Fare'])
```
