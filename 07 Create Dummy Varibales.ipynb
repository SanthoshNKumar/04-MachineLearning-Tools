```python
import pandas as pd
import numpy as np

dict = {"Name":["Santhosh","Anand","Sreedevi",'Prashanth','Sachin'],
        "Age":[32,34,31,30,35],
        "Sex":["Male","Male","Female","Male","Male"],
         "Company":["ABB","Dell","Tesco","Tesco","Volvo"],
         "Grade":["P1","P2","P1","P3","P1"]}

df = pd.DataFrame(dict)

df
```

```python
# Create Dummpy Varibales using .map

df['Sex_Male'] = df['Sex'].map({"Female":0,"Male":1})

# Drop the Colum 'Sex'
df.drop('Sex',axis=1,inplace=True)

# Rename the Column 'Sex_Male' to Sex
df.rename(columns={"Sex_Male":'Sex'})
```

```python
dict = {"Name":["Santhosh","Anand","Sreedevi",'Prashanth','Sachin'],
        "Age":[32,34,31,30,35],
        "Sex":["Male","Male","Female","Male","Male"],
         "Company":["ABB","Dell","Tesco","Tesco","Volvo"],
         "Grade":["P1","P2","P1","P3","P1"]}

df = pd.DataFrame(dict)

df
```

```python
# Alternative method for creating dummy variables
pd.get_dummies(df.Sex)
```

```python
# using iloc works on positions (integers)
pd.get_dummies(df.Sex, prefix='Sex').iloc[:, 1:]
```

```python
# alternative using loc that works on labels
pd.get_dummies(df.Sex, prefix='Sex').loc[:, 'Sex':]
```

```python
# Get the Count of Unique elements in the column
df.Sex.value_counts()
```

```python
# we only need k - 1, hence 2 dummy variables here
pd.get_dummies(df.Sex, prefix='Sex').iloc[:, 1:]
```

```python
# create dummy variables for multiple categories
pd.get_dummies(df,columns=['Sex','Grade'])
```

```python
# create dummy variables for multiple categories along with drop the first categorical column
pd.get_dummies(df,columns=['Sex','Grade'],drop_first=True)
```
