```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Titanic_train.csv")

# Get null Sum Values
values = df.isnull().sum()


nullCount = [x for x in values]

# Plot Chart
plt.figure(figsize=(16,7))

plt.bar(df.columns,nullCount)
plt.grid()
plt.xlabel("Columns")
plt.ylabel("Rows Count")
```
