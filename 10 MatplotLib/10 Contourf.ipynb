```python
import numpy as np
import matplotlib.pyplot as plt

X= np.linspace(-3.0,3.0,3)
y =np.linspace(-3.0,3.0,4)
```

```python
print(X)
print(y)
```

```python
XX,yy = np.meshgrid(X,y)
```

```python
print(XX,'\n')
print(yy)
```

```python
Z = np.sqrt(XX**2 + yy**2)
```

```python
cp = plt.contour(XX,yy,Z)
plt.clabel(cp, inline=True, fontsize=10)
```
