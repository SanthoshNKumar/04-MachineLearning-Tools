```python
import numpy as np
import matplotlib.pyplot as plt

# sample plot
x = np.linspace(-10, 10, 50)


# 2 rows and 2 columns
# figsize = width and height (in inches)

fig, ax_list = plt.subplots(2,2,sharex=True, figsize=(8, 6))

y=np.sin(x)
ax_list[0,0].plot(x,y) 
ax_list[1,0].plot(x,y)

y=np.cos(x)
ax_list[0,1].plot(x,y)
ax_list[1,1].plot(x,y)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("concrete_data.csv")

plt.figure(figsize=(15,10.5))
plot_count = 1

for feature in list(df.columns)[:-1]:
    plt.subplot(3,3,plot_count)
    plt.scatter(df[feature], df['concrete_strength'])
    
    plt.xlabel(feature.replace('_',' ').title())
    plt.ylabel('Concrete strength')
    plot_count+=1
    
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,30))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i in range(1, 5):
    ax = fig.add_subplot(5, 1, i)
    ax.text(0.5, 0.5, str((5, 1, i)),
           fontsize=18, ha='center')
```
